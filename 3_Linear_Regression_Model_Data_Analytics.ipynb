## T-Test

```python
import pandas as pd

cleand_data_csv = 'https://raw.githubusercontent.com/charmy2525/DataAnalytics/main/Data%20analytics%20course%20data%20set%20-%20student%20academic%20performance%20-%20output%20after%20data%20cleaning.csv'
data_file ='https://raw.githubusercontent.com/charmy2525/DataAnalytics/main/Data%20analytics%20course%20data%20set%20-%20student%20academic%20performance%20-%20data.csv'
df = pd.read_csv(cleand_data_csv)
#df = pd.read_csv(data_file)

feature = 'Calculus_1'
rvs1 = df[(df.Year_of_Entry == 2020)]
rvs2 = df[(df.Year_of_Entry == 2019)]
data1 = rvs1[feature].astype(float)
data2 = rvs2[feature].astype(float)
data1.index = pd.RangeIndex(start=0, stop=0+len(data1), step=1)
data2.index = pd.RangeIndex(start=0, stop=0+len(data2), step=1)

fig = plt.figure(figsize = (30,10))
ax = fig.add_subplot(121)
bx = fig.add_subplot(122)

ax.set_title(feature+' Divide by Number of Data',fontweight="bold", fontsize=22,pad=30)
bx.set_title(feature+' Original Data',fontweight="bold", fontsize=22,pad=30)

a_heights, a_bins = np.histogram(data1)
b_heights, b_bins = np.histogram(data2, bins=a_bins)

width = (a_bins[1] - a_bins[0])/3

ax.bar(a_bins[:-1], a_heights/len(data1), width=width, facecolor='red')
ax.bar(b_bins[:-1]+width, b_heights/len(data2), width=width, facecolor='black')

bx.bar(a_bins[:-1], a_heights, width=width, facecolor='cornflowerblue')
bx.bar(b_bins[:-1]+width, b_heights, width=width, facecolor='seagreen')

from scipy import stats
alpha=0.05

stats.ttest_ind(data1, data2,equal_var=False)
pv=stats.ttest_ind(data1, data2).pvalue
print('p-value = ',pv)
if (stats.ttest_ind(data1, data2).pvalue>=alpha) :
  {
      print("can not reject H0 : Two data is no significant different.")
  }
else :
  {
      print("reject H0 : Two data is significant different.") 
  }
```

## Single Variable Linear Regression

```python
import numpy as np

data_X = df[features_x]
data_Y = df[features_y]

# Split the targets into training/testing sets
y_train = data_Y[int(len(data_Y)*0.2)+1:]
y_test = data_Y[0:int(len(data_Y)*0.2)]

x_train = data_X[int(len(data_Y)*0.2)+1:]
x_test = data_X[0:int(len(data_Y)*0.2)]

#convert to array
x_train_arr = np.array(x_train)
y_train_arr = np.array(y_train)
x_test_arr = np.array(x_test)
y_test_arr = np.array(y_test)
```

## Linear Regression Line

```python
# Create linear regression object
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train_arr[:, np.newaxis], y_train_arr)

# The coefficients
print("Coefficients: \n", lm.coef_[0])
print("Interception: \n",lm.intercept_)
print("equation: "+features_y+" = ",lm.coef_[0],"*"+features_x+ " + ",lm.intercept_)


# Make predictions using the testing set
y_pred = lm.predict(x_test_arr[:, np.newaxis])

# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % lm.score(x_test_arr[:, np.newaxis], y_test_arr))
Adj_r2 = 1 - (1-lm.score(x_test_arr[:, np.newaxis], y_test_arr)) * (len(y_test_arr)-1)/(len(y_test_arr)-x_test_arr.ndim-1)
print("R2-Adj: %.2f" % Adj_r2)
# Plot outputs
plt.figure(figsize = (20,10))
plt.scatter(data_X,data_Y,s=area*5)
plt.plot(x_test_arr, y_pred,color="red");
plt.xlabel(features_x)
plt.ylabel(features_y)
plt.show()
```

## Multiple Variable Linear Regression

```python
# Split the targets into training/testing sets
import numpy as np

features_x = ['Calculus_1','Physics']
features_y = 'University_GPAx_1st_year'
data_X =df[features_x]
data_Y = df[features_y]

data_X = data_X.astype(float)
data_Y = data_Y.astype(float)

x_train = data_X[int(len(data_Y)*0.2)+1:]
x_test = data_X[0:int(len(data_Y)*0.2)]
y_train = data_Y[int(len(data_Y)*0.2)+1:]
y_test = data_Y[0:int(len(data_Y)*0.2)]

x_train_arr = np.array(x_train)
y_train_arr = np.array(y_train)
x_test_arr = np.array(x_test)
y_test_arr = np.array(y_test)

from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(x_train_arr, y_train_arr)
# The coefficients
print("Coefficients: \n", lm.coef_)
print("Interception: \n",lm.intercept_)

equation = "equation: "+features_y+" = "
equation_round = "equation_round: "+features_y+" = "

for i in range(len(data_X.columns)):
  equation = equation+str(lm.coef_[i])+"*"+data_X.columns[i]
  equation_round = equation_round+str(round(lm.coef_[i],2))+"*"+data_X.columns[i]
  try:
    if lm.coef_[i+1] > 0 :
      equation = equation+"+"
      equation_round = equation_round+"+"
  except:
    pass
if lm.intercept_ > 0 : 
  equation = equation+"+"
  equation_round = equation_round+"+"
equation = equation+str(lm.intercept_)
equation_round = equation_round+str(round(lm.intercept_,2))
print(equation)
print(equation_round)

# Make predictions using the testing set
y_pred = lm.predict(x_test_arr)

# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % lm.score(x_test_arr, y_test_arr))
Adj_r2 = 1 - (1-lm.score(x_test_arr, y_test_arr)) * (len(y_test_arr)-1)/(len(y_test_arr)-x_test_arr.ndim-1)
print("R2-Adj: %.2f" % Adj_r2)

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

X = x_test_arr[:,0]
Y = x_test_arr[:,1]
X, Y = np.meshgrid(X, Y)
Z = lm.coef_[0]*X-lm.coef_[1]*Y+lm.intercept_
fig = plt.figure(figsize = (20,15))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot', linewidth=0, antialiased=False)
ax.set_xlabel(data_X.columns[0])
ax.set_ylabel(data_X.columns[1])
ax.set_zlabel(data_Y.name)
ax.view_init(30, 45)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
#Label Encoding
import pandas as pd
cleand_data_csv = 'https://raw.githubusercontent.com/charmy2525/DataAnalytics/main/Data%20analytics%20course%20data%20set%20-%20student%20academic%20performance%20-%20output%20after%20data%20cleaning.csv'
data_file ='https://raw.githubusercontent.com/charmy2525/DataAnalytics/main/Data%20analytics%20course%20data%20set%20-%20student%20academic%20performance%20-%20data.csv'
df = pd.read_csv(cleand_data_csv)

non_numeric_columns = pd.Index(set(df.columns).difference(set(df.select_dtypes(include=(np.number)).columns)))

df_label = df.copy()

from sklearn import preprocessing
lable_encoding = preprocessing.LabelEncoder()
for col in non_numeric_columns:
  lable_encoding.fit(df[col])
  label=lable_encoding.transform(df[col])
  df_label = pd.concat([df_label, pd.DataFrame(label,columns = [col+'_label'])], axis = 1)
df_label
```
